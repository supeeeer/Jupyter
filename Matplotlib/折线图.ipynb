# 折线图

````python
# from __future__ import print_function, division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import random
# %matplotlib inline
````

````python
matplotlib.rcParams['figure.figsize'] = [20, 8]
matplotlib.rcParams['figure.dpi'] = 120
````

## 绘制连续折线图
1. 初始化数据
2. 绘制：.ploy(x,y)
    - `line = plt.plot()`
3. 显示：.show()



````python
x = range(2, 9, 2)
y = [15, 25, 10, 20]
plt.plot(x, y)
plt.show()
````

## 设置图形显示

### 创建figure实例
- plt.figure()：创建一个图形实例

     ```
    matplotlib.pyplot.figure(
        num=None, figsize=None, dpi=None, facecolor=None,
        edgecolor=None, frameon=True, FigureClass=Figure, clear=False, ** kwargs)
    ```
    - num：指定创建的figure名称
    - figsize：以英寸为单位，用元组表示(figsize=(15,3))
    - dpi：图形分辨率
    - facecolor：背景颜色
    - edgecolor：边框颜色
    - frameon：默认值True为绘制边框
    - FigureClass：可以选择使用自定义图形实例
    - clear：重建figure实例

````python
plt.figure(num=1, figsize=(10, 5), dpi=80 )
plt.plot(x, y)
plt.show()
````

### 设置轴

#### 修改轴刻度
- xsticks(list)：显示数字
- xsticks(num_list, str_list)：显示字符串
- xsticks(rotation=90)：旋转

````python
# range步长只能是整数，要创建浮点型，可以创建一个循环列表：
lable = [i/2 for i in range(4, 19)]
print(lable)
print(lable[::3])    # 取整个列表，步长为3

plt.xticks(lable)
plt.plot(x, y)
plt.show()
````

````python
plt.figure(figsize=(20, 8), dpi=80)
y = [random.randint(20, 35) for i in range(120)]

# 保证x和x_lable有想相同的长度
x = [i for i in range(120)]
x_lable = ['11点{}分'.format(i) for i in range(60)] + ['12:点{}分'.format(i) for i in range(60)]
plt.xticks(x[::10], x_lable[::10])

plt.plot(x, y)
plt.show()
````

````python
# 旋转
plt.figure(figsize=(20, 8), dpi=80)
plt.xticks(x[::10], x_lable[::10], rotation=45)
plt.plot(x, y)
plt.show()
````

````python
plt.xticks(())    # 隐藏坐标轴数据
plt.yticks(())
plt.show()
````

#### 获取四条轴
ax = plt.gca()： ‘get current axis’

ax.spines[ 'right' ]： 右边的轴

    ax.spines['right'].set_color('none')：透明
    
    ax.spines['bottom'].set_position(('data', y))：将bottom对准y轴的y

````python
ax = plt.gca()
ax.spines['right'].set_color('none')
plt.show()
````

````python
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0.2))    # 调整x底轴
ax.spines['left'].set_position(('data', 0.2))      # 调整y左轴
plt.show()
````

#### 次坐标轴

`ax2 = ax1.twinx()`：创建一个共享x轴的新轴域

````python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4],'b--')
plt.ylabel('blue one', color='b')

ax2 = plt.twinx()
ax2.plot([1, 2, 3, 4], [4, 3, 2, 1], 'g')
ax2.set_ylabel('green one', color='g')

plt.show()
````

### 注释
plt.annotate()
- s：string
- xy：被注释的点位置
- xytext：text放置的位置
- arrowprops：加箭头

````python
# 源数据
xx = np.linspace(-3, 3, 50)
yy = 3*xx + 1

# 调整坐标轴
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.plot(xx, yy)

# 添加点
x0 = 1
y0 = 3*x0 +1
plt.scatter(x0, y0, s=80, c='k', zorder=3)      # zorder设置层叠关系
plt.plot([x0, x0], [0, y0], 'k--', lw=2.5)      # 'k--' 表示black + ’--‘样式

# 注释
plt.annotate(s='aaaa', xy=(x0, y0), xytext=(2, 3), 
             arrowprops=dict(facecolor='black', shrink=0.09))


plt.show()
````

### 双线
- 多次使用plt.plot()
     - c = ' green ' ：设置线条颜色
     - linestyle = '--' ：设置线条风格
     -  ( color和linestyle可以简化组合使用，如' b-- '表示蓝色 + ’ -- ‘ ) 样式
     - linewidth = 5 ：设置线条粗细
    

````python
y1 = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 4, 3, 1, 1, 1, 1]
y2 = [2, 3, 3, 1, 1, 2, 5, 2, 2, 4, 4, 5, 7, 2, 1, 5, 2, 1, 3]
x = range(11, 30)
plt.plot(x, y1, linestyle='--')
plt.plot(x, y2, linewidth=5)
plt.show()
````

#### 图例
1. plt.plot( x, y, label='...' )
2. plt.legend( loc=,)

````python
plt.plot(x, y1, label='小明', c='green')
plt.plot(x, y2, label='小红', color='#00FFFF')
plt.legend(loc='best')

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
````

### 设置style
style文件路径：/Users/supeeeer/anaconda3/lib/python3.7/site-packages/matplotlib/mpl-data/stylelib

- <span class="mark">matplotlib.rcParams[ '...' ]：配置全局style</span>
    - matplotlib.rcParams.keys()：查看
- plt.style.use( ['...', '...'] )：使用style文件，后者覆盖前者
    - plt.style.available：查看
- with plt.style.context( '...' )：局部使用style

````python
# 设置各个属性
# print(matplotlib.rcParams.keys())    # 查看可以配置的属性
````

````python
matplotlib.rcParams["lines.marker"] = "o"   #所绘制的线将带有圆形的点标识符
matplotlib.rcParams["figure.figsize"] = [20, 8]
matplotlib.rcParams['font.sans-serif'] = ['Arial Unicode MS']

plt.plot(x, y)
plt.show()
````

````python
# 使用原生style
print(plt.style.available)    #查看可以使用的style

plt.style.use(['dark_background', 'seaborn-pastel'])
plt.plot(x, y)
plt.show()
````

**dark_background.mplstyle文件内容:**

lines.color: white

patch.edgecolor: white

text.color: white

axes.facecolor: black
...

因此也可以自定义.mplstyle文件，放在style文件中，在程序中进行调用

````python
# 局部使用
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
    plt.show()
````

````python
# 改回原始配置
plt.style.use('default')
````
