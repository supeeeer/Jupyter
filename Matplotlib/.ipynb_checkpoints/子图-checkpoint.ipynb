````python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
````

# 子图
一个Figure对象可以包含多个子图（Axes），axes可以在figure上画图

生成子图（subplot）的时候，必然带着一套轴域（Axes）。而用轴域（Axes）方法，客观上就是生成了一个可以画图的子图。

- figure：轴域 + 边界 (画板)
- axes = subplot：轴域，整个子图 (画纸)
- axis：一条轴 (axisX、axisY)


## subplots
subplots: 创建一组子图，返回元组。
根据ax定位子图，<span class="mark">可以共享轴</span>

- `(fig, ax) = plt.subplots(nrows, ncols, index)  `

    ```
    matplotlib.pyplot.subplots(
        nrows=1, ncols=1, sharex=False, sharey=False,
        squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
    ```

    - nrows：图例的行数
    - ncols：图例的列数
    - index：子图的索引
    - sharex：控制x轴之间的属性共享
    - facecolor：子图的背景色
    - palor：指示子绘图应该是极坐标。默认为False

````python
xx = np.linspace(-3, 3, 50)
yy = 2*xx + 1
plt.plot(xx, yy)

fig, ax = plt.subplots(2,2)    # 返回4个子图
# fig, ((ax11, ax12), (ax21, ax22)) = plt.subplots(2,2)    # 将ax格式化
ax[0, 0].plot(xx, yy)          # 第一个子图画图

plt.show()
````

````python
plt.subplots(2,2, sharey='row')    # 每行子图共享y,只在第一列处显示y
plt.show()
````

## subplot
subplot：创建子图，并定位到其中一个，返回定位的子图

多行列时，以最小单位计算每个子图的位置

`axes = plt.subplot(nrows, ncols, index, **kwargs) `       
`axes = plt.figure().add_subplot(nrows, ncols, index, **kwargs)`

`axes.set()    # 设置子图坐标轴`


````python
fig = plt.figure()
ax = fig.add_subplot(221)

ax.set(xlim=[0.5, 4.5], ylim=[-2, 8], 
       title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')
plt.show()
````

````python
plt.subplot(221)                           # 将原图分成2*2，并定位到第一个
plt.subplot(222, projection='polar')
plt.subplot(212)
plt.show()
````

## axes

`plt.axes(rect, projection=None, polar=False, **kwargs)`

- rect = [ left, bottom, width, height ] = 起始点 + 大小
- projection : {None, 'aitoff', 'hammer', 'lambert', 'mollweide', 'polar', 'rectilinear', str}, optional

````python
plt.axes([0.1, 0.1, 0.8, 0.8])
plt.axes([0.72, 0.72, 0.16, 0.16]) 
plt.show()
````

## subplot2grid

`ax = plt.subplot2grid(shape, loc, rowspan=1, colspan=1, fig=None, **kwargs)`

- shape：整个图的大小
- loc：当前ax的位置，从(0，0)开始

````python
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1 title')

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
ax6 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.show()
````
