# 图形

- 折线图：同一数据的变化情况
    - `plt.plot(x, y)`

- 直方图：连续数据的分布情况
    - `plt.bar(x, y)`
    - `plt.hist(num, bins)`

- 条形图：离散数据的分布情况
    

- 散点图：数据间的分布情况
    - `plt.scatter(x, y)`


````python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
````

````python
matplotlib.rcParams['figure.figsize'] = [20, 8]
matplotlib.rcParams['font.family'] = 'Arial Unicode MS'
````

## 散点图


### 绘制
`plt.scatter(x, y)`

````python
x = range(11, 30)
y = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 4, 3, 1, 1, 1, 1]
plt.scatter(x, y)
plt.show()
````

### 多组对比
注意要<span class="mark">整体移动图像</span>

````python
y1 = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 4, 3, 1, 1, 1, 1]
y2 = [2, 3, 3, 1, 1, 2, 5, 2, 2, 4, 4, 5, 7, 2, 1, 5, 2, 1, 3]
x1 = range(19)
x2 = range(40, 59)    # 整体移动x轴

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
````

````python
# x轴显示特定部分
# xticks( list ) 需要传入列表参数
y1 = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 4, 3, 1, 1, 1, 1]
y2 = [2, 3, 3, 1, 1, 2, 5, 2, 2, 4, 4, 5, 7, 2, 1, 5, 2, 1, 3]
x1 = range(19)
x2 = range(40, 59)

# 设置x轴显示
plt.xticks(list(x1)+list(x2))

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
````

### 彩色图

````python
X = np.random.normal(size = 1024)
Y = np.random.normal(size = 1024)
C = np.arctan2(Y, X)

plt.scatter(X, Y, s=75, c=C, alpha=.5)

plt.xticks(())    # 隐藏坐标轴数据
plt.yticks(())
plt.show()
````

## 直方图

### 绘制
- `plt.bar()`：多个间比较 ，需要统计后的数据
- `plt.hist()`：单个比较 ，需要原始数据(自动分组)
    - ( histogram直方图 )

````python
# 准备数据
a = ['猩球崛起', '敦刻尔克', '蜘蛛侠', '战狼']
b = [15746, 312, 4497, 319]

# 画图
x = [i for i in range(len(a))]
plt.bar(x, b)
plt.show()
````

### 多组对比
1. 定好宽度
2. 确定第一类的位置
3. 其他类在此基础上向右偏移

````python
# 数据
a = ['猩球崛起', '敦刻尔克', '蜘蛛侠', '战狼']
b1 = [15746, 312, 4497, 319]
b2 = [12357, 156, 2045, 168]
b3 = [2358, 399, 2358, 362]

# 绘制
bar_width = 0.2                      # 宽度限定在（1/条数）范围内，否侧覆盖
x1 = list(range(len(a)))             # x是int_list型的
x2 = [i+bar_width for i in x1]       # x2在x1的基础上右移bar_width个单位
x3 = [i+bar_width*2 for i in x1] 

# 画
plt.bar(x1, b1, width=bar_width)
plt.bar(x2, b2, width=bar_width)
plt.bar(x3, b3, width=bar_width)
plt.show()
````

#### 设置x轴
选择中间处显示名字

````python
plt.xticks(x2, a)

plt.bar(x1, b1, width=bar_width)
plt.bar(x2, b2, width=bar_width)
plt.bar(x3, b3, width=bar_width)
plt.show()
````

#### 设置标签

````python
plt.bar(x1, b1, width=bar_width, label='9月10日')
plt.bar(x2, b2, width=bar_width, label='9月11日')
plt.bar(x3, b3, width=bar_width, label='9月12日')

plt.legend()

plt.show()
````

#### 上下两组
x相同，y不同

````python
n = 12
X = np.arange(n)
Y1 = np.random.uniform(0.5, 1, n)
Y2 = np.random.uniform(0.5, 1, n)

plt.bar(X, Y1, edgecolor='white')
plt.bar(X, -Y2, edgecolor='white')
plt.show()
````

### 单组分布
- `plt.hist(数据, bins)`

    - matplotlib.pyplot.hist(  
    x, bins=10, range=None, normed=False,   
    weights=None, cumulative=False, bottom=None,   
    histtype=u'bar', align=u'mid', orientation=u'vertical',   
    rwidth=None, log=False, color=None, label=None, stacked=False,   
    hold=None, **kwargs)  

- 组数 = 极差[ max(a) - min(a) ] / 组距

````python
a = np.random.randint(90, 180, 120)
plt.hist(a, 20)    # 将a分成20组（20条）
plt.show()
````

#### 频率分布
`plt.hist(density = True)`

````python
a = np.random.randint(90, 180, 120)
plt.hist(a, 20, density=True, rwidth=0.95)    
plt.show()
````

#### 横向图
`plt.hist(orientation='horizontal')`

````python
a = np.random.randint(90, 180, 120)

plt.hist(a, 20,orientation='horizontal')

plt.grid()
plt.show()
````

#### 设置x轴
使用`plt.xticks()`，调整x轴显示与数量

````python
interva = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 60, 90]
widths = [5, 5, 5, 5, 5, 5, 5, 5, 5, 15, 30, 60]
quantity = [836, 2737, 3723, 3976, 3579, 1423, 4472, 643, 864, 653, 210, 47]

plt.bar(range(len(quantity)), quantity, width=1)      # 单纯定义数量（原始数据）

_x = [i-0.5 for i in range(len(quantity)+1)]          # 调整x显示数字
_xlabel = interva + [150]                             # 加上最后一个点
plt.xticks(_x, _xlabel )

plt.grid()
plt.show()
````

### 添加每组数据
- `for a, b, c in zip(A, B, C)`：将ABC打包成元组，使循环同时传入多个变量
- `plt.text( x, y, string )`：输出文字在 ( x, y ) 处

````python
n = 12
X = np.arange(n)
Y1 = np.random.uniform(0.5, 1, n)
Y2 = np.random.uniform(0.5, 1, n)

plt.bar(X, Y1)
plt.bar(X, -Y2)

# 循环添加数据
for x, y1, y2 in zip(X, Y1, Y2):
    plt.text(x-0.1, y1+0.05, '{:.2f}'.format(y1))
    plt.text(x-0.1, -y2-0.05, '{:.2f}'.format(-y2))

plt.show()
````

## 色彩图形

1. `plt.imshow()`

    `image = matplotlib.pyplot.imshow(X, cmap=None, norm=None,
    aspect=None, alpha=None, vmin=None, vmax=None,
    origin=None, extent=None,filternorm=1, filterrad=4.0,data=None, **kwargs)`

        - X：要绘制的图像, 二维数组
        - cmap：颜色图谱（colormap), 默认绘制为RGB(A)颜色空间
        - norm：将数组X映射到colormap
        - interpolation：各个像素间的关系
    
2. `plt.colorbar()`

````python
a = np.linspace(0.3, 0.7, 9).reshape(3, 3)

plt.imshow(a, interpolation='nearest', cmap='bone')
plt.colorbar()
plt.show()
````

````python
a = np.linspace(0.3, 0.7, 9).reshape(3, 3)
fig, ax = plt.subplots(2, 2)

# 创建子图
img = ax[0, 0].imshow(a, interpolation='nearest', cmap='bone')     # 返回img对象
ax[0, 1].imshow(a, interpolation='nearest')
ax[1, 0].imshow(a, interpolation='gaussian', cmap='bone')
ax[1, 1].imshow(a)

fig.colorbar(img, ax=ax[0, 0])                                    # 传入img对象+ax轴域

plt.show()
````

## 3D图像
1. 将一维X, Y拟合成二维数组 ( Z需要传入二维数组 )
`np.meshgrid())`
2. 构建figure，构建axes
3. ax.plot_surface()

`ax.plot_surface(X, Y, Z)`
- rstride：宽度步长
- cstride：长度步长

````python
from mpl_toolkits.mplot3d import Axes3D
````

````python
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)       # 构建坐标轴网格,二维
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure() 
ax = Axes3D(fig)               # ax作为3D图像
ax.plot_surface(X, Y, Z, cmap='rainbow')  

plt.show()
````

### 子图
```
ax = fig.add_subplot(221, projection='3d')
ax.plot_surface()
```

````python
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)   
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(20,15))  # 画板

ax = fig.add_subplot(221, projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow', rstride=1, cstride=1)    # 在ax轴域画图

ax = fig.add_subplot(222, projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow', rstride=5, cstride=1)    

plt.show()
````

### 投影图

`contour()` , `contourf()` 用于画等高线、等高填充图

- zdir：投影方向

````python
# contour
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)   
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(20,15)) 

ax = fig.add_subplot(221,projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow')  
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='rainbow')      # 画z的投影线
ax.set_zlim(-2, 2)

ax = fig.add_subplot(222,projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow')  
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap='rainbow')     # 画z的投影图
ax.set_zlim(-2, 2)

ax = fig.add_subplot(223,projection='3d')
ax.plot_surface(X, Y, Z, cmap='rainbow')  
ax.contourf(X, Y, Z, zdir='x', offset=-4, cmap='rainbow')      # 画x的投影图

plt.show()
````
