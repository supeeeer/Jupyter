# Robots协议
- 是什么

    网络爬虫排除标准（Robots Exclusion Protocol），用来告诉爬虫和搜索引擎哪些页面可以抓取，哪些不可以

    它通常是一个 robots.txt 文本文件，放在网站的根目录下，和网站的入口文件（比如 index.php、index.html 和 index.jsp 等）放在一起

- 怎么做

    当爬虫访问一个站点时，会检查这个站点根目录下是否存在 robots.txt 文件
    
    如果存在，爬虫会根据其中定义的爬取范围来爬取。如果没有，就访问所有可直接访问的页面
    
- 实例
    ```
    # 对 ’Baiduspider‘ 爬虫，只允许爬取 public 目录
    User-agent: Baiduspider         # 爬虫名称
    Disallow: /                     # 不允许爬取的目录
    Allow: /public/                 # 允许爬取的目录
    
    # 对所有爬虫，禁止爬取/private/和/tmp/目录
    User-agent: *  
    Disallow: /private/  
    Disallow: /tmp/
    
    # 只允许 ’WebCrawler‘ 爬虫，爬取所有页面
    User-agent: WebCrawler  
    Disallow:  
    User-agent: *  
    Disallow: /
    ```

````python
from urllib.request import urlopen

robot = urlopen('https://www.zhihu.com/robots.txt')
print(robot.read().decode('utf8'))
````

# robotparser模块
使用 robotparser 模块提供的 **RobotFileParser类** 来解析 robots.txt ，判断一个爬虫是否有权限来爬取这个网页

## RobotFileParser类
`urllib.robotparser.RobotFileParser(url='')`

他包含的方法有：
- `set_url()` ：传入 robots.txt 文件的url，也可以在初始化时就设置
- `read()`：读取并解析 robots.txt ，不会返回任何内容，但如果不调用这个方法，接下来的判断都会为 False
- `parse(str)`：传入 robots.txt  (某些行) 的内容，并解析，不需要调用`read()`
    - <span class="mark">❗️urlopen() 得到 HTTPResponse 对象 —— read() 得到 bytes —— decode() 得到 str</span>
- `can_fetch(User-agent, URL) = bool`：判断该搜索引擎能否抓取这个 URL
- `mtime()`：返回上次抓取到 robots.txt 的时间，对于长时间抓取的爬虫，需要定期检查来抓取最新的 robots.txt
- `modified()`：修改 `mtime()` = 当前时间

````python
from urllib.robotparser import RobotFileParser

rp = RobotFileParser('https://www.zhihu.com/robots.txt')
# rp.set_url('https://www.zhihu.com/robots.txt')
rp.read()

print(rp.can_fetch('Googlebot', 'https://www.zhihu.com'))
print(rp.can_fetch('Googlebot', 'https://www.zhihu.com/search'))
print(rp.mtime())
````
