## urlopen() 
```
HTTPResponse = urllib.request.urlopen(url, data=None, timeout, cafile=None, capath=None, context=None)
             = urllib.request.urlopen(Request)
```
- data：需要上传bytes类型。<span class="mark">使用 bytes() 将参数转化为字节流格式，同时请求方法变为 POST (URLopen默认 GET 上传)</span> 
- timeout：设置超时时间，单位为秒
- context：必须是 ssl.SSLContext 类型，用来指定 SSL 设置
- cafile 和 capath：分别指定 CA 证书和它的路径，在请求 HTTPS 链接时会有用


### HTTPResponse 对象

````python
import urllib.request

# 属性
response = urllib.request.urlopen('https://www.python.org')    
print(response.msg,         # 响应头实例
      response.status,      # 状态码
      response.version,     # HTTP协议版本号,11 = HTTP/1.1
      response.reason,      # 响应描述字符串
      response.closed,      # 是否关闭
      response.debuglevel)
````

````python
# 方法
print(response.info())                     # 响应头信息,返回字符串
print(response.getheaders(),'\n')          # 整个响应头,返回列表
print(response.getheader('Server'),'\n')   # 响应头的一个信息

print(response.geturl())         # 实际请求的url(防止重定向)
print(response.getcode())        # 状态码
````

<span class="burk">❓❓response.read()之后，无法再使用response.read()</span>

<span class="mark">❗️read()得到未解码的数据 (bytses) ，需要配合decode()</span>

````python
response = urllib.request.urlopen('https://www.python.org')    
print(response.read().decode('utf-8'))
````

### 请求包含data

````python
import urllib.request
import urllib.parse

data = {'word': '我 hello'}
print(data, type(data))

data = urllib.parse.urlencode(data)    # dict → string
print(data, type(data))

data = bytes(data, encoding='utf8')    # string → bytes
print(data, type(data), '\n\n')


response = urllib.request.urlopen('https://httpbin.org/post', data=data)
print(response.read().decode('utf-8'))
````

### 设置TimeOut 

利用` try except `语句，一个网页如果长时间未响应，就跳过它的抓取

其异常属于` urllib.error `模块

````python
import socket  
import urllib.request  
import urllib.error  

try:
    response = urllib.request.urlopen('http://httpbin.org/get', timeout=0.1)
except urllib.error.URLError as e:
    if isinstance(e.reason, socket.timeout):
        print('TIME OUT')
````

## Request类 - request 
构建一个完整的请求，在请求中加入 Headers 等信息

    class urllib.request.Request(
        url, 
        data=None, 
        headers={}, 
        origin_req_host=None, 
        unverifiable=False, 
        method=None)
    
- url：请求的 URL，必传参数，其他都是可选参数
- data：必须传 bytes（字节流）类型
- headers：字典，它就是请求头，可以直接构造，或通过 `add_header()` 
添加
    - 默认的 User-Agent 是 Python-urllib，我们可以修改它来伪装浏览器
- origin_req_host：请求方的 host 名称或者 IP 地址
- unverifiable：表示这个请求是否是无法验证的，默认是 False，表示用户有权限来选择接收这个请求的结果
    - 例如，我们请求一个 HTML 文档中的图片，但是我们没有自动抓取图像的权限，这时 unverifiable = True
- method：字符串，用来指示请求使用的方法，比如 GET、POST 和 PUT 


````python
from urllib import request, parse

url = 'https://httpbin.org/post'
headers = {
    'User-Agent': 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)',  
    'Host': 'httpbin.org'  
}
data_dict = {'name': 'Germey'}
data = bytes(parse.urlencode(data_dict), encoding='utf8')

req = request.Request(url=url, data=data, headers=headers, method='post')
req.add_header('User-Agent', 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)')

response = request.urlopen(req)
print(response.read().decode('utf-8'))
````

## Opener
Opener对比URLopen更底层，可以实现更高级的功能。

## Handler
各种处理器，处理登录验证、处理 Cookies 、处理代理设置

- BaseHandler 类，是其他 Handler 的父类，它提供了最基本的方法，例如 default_open、protocol_request 
- HTTPDefaultErrorHandler 用于处理 HTTP 响应错误，错误都会抛出 HTTPError 类型的异常。
- HTTPRedirectHandler：处理重定向
- HTTPCookieProcessor：处理 Cookies
- ProxyHandler：设置代理，默认代理为空
- HTTPPasswordMgr：密码管理，为一个域指定用户名和密码
    - HTTPPasswordMgrWithDefaultRealm：不指定域，直指定url
- HTTPBasicAuthHandler：管理认证，比如一个链接打开时需要认证

### 认证管理
使用`HTTPPasswordMgr`配合`HTTPBasicAuthHandler`完成指定URL的用户名密码映射

❗️`HTTPPasswordMgr - HTTPBasicAuthHandler - opener`

````python
from urllib.request import HTTPPasswordMgrWithDefaultRealm, HTTPBasicAuthHandler, build_opener  
from urllib.error import URLError  

username = 'username'  
password = 'password'  
url = 'http://localhost:5000/'  

p = HTTPPasswordMgrWithDefaultRealm()               # 管理密码的对象
p.add_password(None, url, username, password)       # URL 与 (账号，密码) 映射, 域为none
auth_handler = HTTPBasicAuthHandler(p)              # HTTPPasswordMgr → HTTPBasicAuthHandler

opener = build_opener(auth_handler)                 # HTTPBasicAuthHandler → opener

try:  
    result = opener.open(url)                       # response对象
    html = result.read().decode('utf-8')            # 解码后的文件
    print(html)  
except URLError as e:  
    print(e.reason)
````

### 设置代理
`ProxyHandler({})` 参数是一个字典，键名是协议类型（比如 HTTP 、 HTTPS 等），键值是代理链接，可以添加多个代理

````python
from urllib.request import ProxyHandler, build_opener 
from urllib.error import URLError 

proxy = ProxyHandler({ 
    'http': 'http://127.0.0.1:9743',                      # 在本地搭建了一个代理，运行在 9743 端口
    'https': 'https://127.0.0.1:9743' 
})

opener = build_opener(proxy)

try:
    response = opener.open('https://www.baidu.com')       # 必须加上https
    print(response.readline())
except URLError as e:
    print(e.reason)
````

### cookie
使用 `CookieJar` 配置 `HTTPCookieProcessor` ，得到的cookie存放在 `CookieJar` 中 

❗️(`CookieJar` - `HTTPCookieProcessor` - `opener`)

在保存 / 读取中，需要使用同一类的 `CookieJar` ，并且配置相同的参数

````python
# 获取
from http import cookiejar
from urllib.request import HTTPCookieProcessor, build_opener
from urllib.error import HTTPError

cookies = cookiejar.CookieJar()                         
cookie_handler = HTTPCookieProcessor(cookies)           # CookieJar → HTTPCookieProcessor
opener = build_opener(cookie_handler)

try:
    response = opener.open('http://www.baidu.com')
    for cookie in cookies:                              # cookie 存放在 CookieJar 中
        print(cookie.name+' = '+cookie.value)
except HTTPError as e:
    print(e.reason)
````

````python
# 保存
filename = 'cookies.txt'
cookies = cookiejar.MozillaCookieJar(filename)         # 以MozillaCookieJar形式进行保存
cookie_handler = HTTPCookieProcessor(cookies)
opener = build_opener(cookie_handler)
try:
    response = opener.open('http://www.baidu.com')
    cookies.save(ignore_discard=True)                  # 保存，设置保存形式
except HTTPError as e:
    print(e.reason)
````

````python
# 读取
filename = 'cookies.txt'
cookies = cookiejar.MozillaCookieJar()
cookies.load(filename, ignore_discard=True)            # 读取，注意和保存形式相同
cookie_handler = HTTPCookieProcessor(cookies)
opener = build_opener(cookie_handler)
try:
    response = opener.open('http://www.baidu.com')
    print(response.readline())
except HTTPError as e:
    print(e.reason)
````
