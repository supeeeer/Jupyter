# PIL (pillow)
https://pillow.readthedocs.io/en/3.1.x/reference/Image.html#functions

```python
from PIL import Image
```

## Image属性
- format：格式
    - https://github.com/python-pillow/Pillow/blob/aaca672173413883fbcefd659f04d74fe44fb5d5/docs/handbook/image-file-formats.rst
- mode：风格。“1”, “L”, “RGB”, or “CMYK”
- size：2-tuple (width, height)

## Image操作
- 保存：`Image.save(file_name, format=None, **params)`
    - format = ‘JPG’：图像格式。也可以在name中写出
    - quality = 95：最大值
    
    
- 获取对应坐标的像素值：`img.getpixel(x, y)`

### 转换为数组
- 图像 → 数组：np.array()
- 数组 → 图像：Image.fromarray()、Image.frombytes()、Image.fromstring()

```python
# 转换成数组
import numpy as np

img = Image.open('img/convert.jpg')
img_mat = np.array(img)
print(img_mat.shape)
print(img_mat[:, :, 0])
```

### histogram()：每个像素值的点个数
`list = Image.histogram(mask=None, extrema=None)`

- 列表下标 = 像素值

- 列表值 = 对应点个数

```python
print(img.convert('1').histogram(),'\n')

for i,k in enumerate(img.convert('1').histogram()):
    if k>10:
        print(i, k)    # 像素值 数量
```

### convert(): 图片模式改变
`image = Image.convert(mode)`

mode:

- 1：只有黑(0)、白(255)两者

- L：灰度，[0， 255]表示不同的灰度

- RGB：3x8位像素

- RGBA：4x8位像素，有透明通道的真彩色

- CMYK：4x8位像素
    - PNG模式的无法使用CMYK保存

- YCbCr：3x8位像素，彩色视频格式

```python
img = Image.open('img/convert.jpg')
print(img.getpixel((450, 200)))
img
```

```python
img1 = img.convert('1')
print(img1.getpixel((450, 200)))
img1
```

```python
imgL = img.convert('L')
print(imgL.getpixel((450, 200)))
imgL
```

```python
imgA = img.convert('RGBA')
print(imgA.getpixel((450, 200)))
imgA
```
