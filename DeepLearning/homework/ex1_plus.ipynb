## z-score标准化
z-score标准化：适用于属性的最大值和最小值未知，或有超出取值范围的离群数据的情况。

标准差：数据集和平均值的差异大小

$$z=\frac{x-\bar{x}}{σ}$$

1. data - data.mean()
    - 新数据的平均值 = 0
    - 下移
2. / data.std()
    - 新数据的标准差 = 1
    - 压扁
    - 一个数据，距离平均数多少个标准差
    
    
<span class="mark">我们不需要把这一步应用到x0中，因为x0总是等于1的，所以它不可能有为0的的平均值</span>

```python
print(data.mean(),'\n')
print(data.std())
data = (data - data.mean()) / data.std()
data.head()
```

```python
ax = data.plot(kind='scatter',x='Price',y='Size',label='Size',figsize=(12,8))
ax2 = ax.twinx()
data.plot(kind='scatter',x='Price',y='Bedrooms',ax=ax2,color='red',label='Bedroom')
plt.show()
```

# 提出模型

## 代价函数
同ex1，<span class="mark">此函数参数个数是可变的</span>

代价函数：$$ J\left( \theta  \right)=\frac{1}{2m}\sum\limits_{i=1}^{m}{\left( {h_\theta}\left( {x^{(i)}} \right)-{y^{(i)}}\right)}^2 $$

其中假设函数：$$ {h_\theta}\left(x \right)={\theta^T}X={\theta_0}+{\theta_1}{x_1} + ... +{\theta_i}{x_i} $$ 

```python
def computeCost(x, y, theta):
    inner = np.power(x * theta.T - y, 2)
    return np.sum(inner)/(len(x)*2)
```

## 梯度下降函数
$${\theta_j}:={\theta_j}-\alpha \frac{\partial }{\partial {\theta_j}}J\left( \theta  \right)$$

```python
def gradientDescent(x, y, theta, alpha, iters):
    cost = np.zeros(iters)
    
    for i in range(iters):
        dtheta = (x * theta.T - y).T * x / len(x)
        theta = theta - alpha * sum(dtheta)
        
        cost[i] = computeCost(x, y, theta)
        
    return theta,cost
```

# 训练参数

## 预处理x,y,θ

```python
data.insert(0,'Ones',1)                   # 添加bias

cols = data.shape[1]                      # 提取x, y
x = data.iloc[:, 0:cols-1]
y = data.iloc[:, cols-1:cols]

x = np.matrix(x)                          # 将x, y转换成matrix，方便操作
y = np.matrix(y)

theta = np.matrix(np.array([0, 0, 0]))    # 初始化参数
```

## 训练

```python
alpha = 0.01
iters = 1000

g, cost = gradientDescent(x, y, theta,alpha,iters)
g2, cost2 = gradientDescent(x, y, theta,0.03,iters)
g3, cost3 = gradientDescent(x, y, theta,0.1,iters)

g4, cost4 = gradientDescent(x, y, theta,alpha,100)
g5, cost5 = gradientDescent(x, y, theta,alpha,1000)
g6, cost6 = gradientDescent(x, y, theta,alpha,2500)

print(g)
```

# 显示

```python
size = np.linspace(data.Size.min(), data.Size.max(), 100)
bedroom = np.linspace(data.Bedrooms.min(), data.Bedrooms.max(), 100)
price = g[0,0]+g[0,1]*size+g[0,2]*bedroom
ax = data.plot(kind='scatter',x='Price',y='Size',label='Size',figsize=(12,8))
ax2 = ax.twinx()
data.plot(kind='scatter',x='Price',y='Bedrooms',ax=ax2,color='red',label='Bedroom')

ax.plot(price,size,'g',label='')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(range(iters),cost,'r',label='0.01')
ax.plot(range(iters),cost2,'b',label='0.03')
ax.plot(range(iters),cost3,'g',label='0.1')

ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('different α')
ax.legend()
plt.show()
```
